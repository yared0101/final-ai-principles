# make a class that takes the generated path, nodes with coordinates and their edges to display agent movement


```python
class GraphModel:
    def __init__(self, nodes, edges, path, show_background_image=True, delay=500):
        self.nodes = nodes
        self.path = path
        self.show_background_image = show_background_image
        self.edges = edges
        self.delay = delay

    def show_path(self, canvas, path):
        # Draw the path
        forward_path_color = "blue"
        backward_path_color = "red"
        path_width = 3
        delay = self.delay  # Delay in milliseconds
        steps_per_line = 10  # Number of steps per line

        def draw_lines(index, step):
            if index < len(path) - 1:
                path_color = forward_path_color
                # to check if this is moving backward, check if the node to go to is in the previous path
                if index > 0 and path[index + 1] in path[:index]:
                    path_color = backward_path_color
                start_node = path[index]
                end_node = path[index + 1]
                start_x, start_y = nodes[start_node]
                end_x, end_y = nodes[end_node]
                x_step_size = (end_x - start_x) // steps_per_line
                y_step_size = (end_y - start_y) // steps_per_line
                current_x = start_x + (step * x_step_size)
                current_y = start_y + (step * y_step_size)
                canvas.create_line(
                    start_x,
                    start_y,
                    current_x,
                    current_y,
                    fill=path_color,
                    width=path_width,
                )
                if step < steps_per_line:
                    canvas.after(delay, draw_lines, index, step + 1)
                else:
                    canvas.after(delay, draw_lines, index + 1, 0)

        # Start drawing the lines with a delay
        draw_lines(0, 0)

    def mouse_motion(self, event, status_bar):
        # Get the coordinates of the mouse cursor
        x = event.x
        y = event.y

        # Update the status bar with the coordinates
        status_bar.config(text=f"Mouse Coordinates: ({x}, {y})")

    def show_nodes(self, canvas):
        # Draw circles on the nodes
        node_radius = 20
        text_font = ("Arial", 12)

        for node in self.nodes:
            x, y = self.nodes[node]
            canvas.create_oval(
                x - node_radius,
                y - node_radius,
                x + node_radius,
                y + node_radius,
                outline="black",  # Set outline color
            )
            if not self.show_background_image:
                canvas.create_text(x, y, text=node, font=text_font)

    def show_edges(self, canvas):
        edge_color = "black"
        edge_width = 2

        for start_node, end_nodes in self.edges.items():
            start_x, start_y = self.nodes[start_node]
            for end_node in end_nodes:
                end_x, end_y = self.nodes[end_node]
                canvas.create_line(
                    start_x, start_y, end_x, end_y, fill=edge_color, width=edge_width
                )

    def create_gui(self):
        # Create the main window
        root = tk.Tk()
        root.title("Traveling Ethiopia Model")

        # Set the dimensions of the window
        width = 1062
        height = 668
        canvas = tk.Canvas(root, width=width, height=height)
        canvas.pack()

        if self.show_background_image:
            # Load the image
            image_path = "Figure 5 Traveling Ethiopia.png"
            image = tk.PhotoImage(file=image_path)
            # Display the image on the canvas
            canvas.create_image(0, 0, anchor=tk.NW, image=image)
        else:
            self.show_nodes(canvas)
            self.show_edges(canvas)

        # Create a status bar
        status_bar = tk.Label(
            root, text="Mouse Coordinates: (0, 0)", bd=1, relief=tk.SUNKEN, anchor=tk.W
        )
        status_bar.pack(side=tk.BOTTOM, fill=tk.X)

        # Bind mouse motion event to update the status bar
        canvas.bind("<Motion>", lambda event: self.mouse_motion(event, status_bar))
        # Run the Tkinter event loop
        self.show_path(canvas, self.path)
        root.mainloop()


# Call the function to create the GUI
graph = GraphModel(
    nodes=nodes,
    edges=edges,
    path=["Addis Ababa", "Adama", "Assela", "Assasa", "Doddolla"],
    show_background_image=True,
    delay=10,
)
graph.create_gui()
```

# re define here our uniformed search class from question

### we will be making 1 change. that is when a node is deemed a deadend, instead of popping from path, push the parent back to the path. so that our agent recognizes it as turning back


```python
class PathFinder:
    def __init__(self, paths, initial_state, goal_state, search_strategy):
        self.paths = paths
        self.initial_state = initial_state
        self.goal_state = goal_state
        self.search_strategy = search_strategy

    def find_path(self):
        if self.search_strategy == "bfs":
            return self.bfs()
        elif self.search_strategy == "dfs":
            return self.dfs()
        else:
            return "Invalid search strategy"

    def bfs(self):
        queue = [[self.initial_state]]
        visited = set()
        while queue:
            # FIFO strategy so pop from the front
            path = queue.pop(0)
            node = path[-1]
            if node == self.goal_state:
                return path
            if node not in visited:
                for neighbour in self.paths[node]:
                    new_path = list(path)
                    new_path.append(neighbour)
                    queue.append(new_path)
                visited.add(node)
        return "No path found"

    def dfs(self):
        path = [self.initial_state]
        visited = set()
        dead_ends = set()
        while path:
            node = path[-1]
            if node == self.goal_state:
                return path
            visited.add(node)
            next_node = None
            for neighbour in self.paths[node]:
                if neighbour not in visited and neighbour not in dead_ends:
                    next_node = neighbour
                    break
            if next_node:
                path.append(next_node)
            else:
                # means we're at a dead end, either a non goal terminal node or a node with all neighbours explored
                # backtrack
                dead_ends.add(node)
                # push back the parent
                # edge case where parent could be a dead end
                negative_len = -len(path)
                start_index = -2
                while path[start_index] in dead_ends:
                    start_index -= 1
                if start_index < negative_len:
                    return "No path found"
                path.append(path[start_index])
        return "No path found"
```

```python
search_strategy = "dfs"
bfs = PathFinder(
    paths=edges,
    initial_state="Addis Ababa",
    goal_state="Kebri Dehar",
    search_strategy=search_strategy,
)
generated_path = bfs.find_path()
print(f"{search_strategy} -", generated_path)
```
